Código en MAXIMA/PYTHON/Python para calcular centroide de un triángulo (Problema 3 Sección 1.1.6).

```python
#DESARROLLO CON SIMPY
#Definir las variables simbólicas
x1, y1, x2, y2, x3, y3 = sp.symbols('x1 y1 x2 y2 x3 y3')

#Coordenadas de los vértices del triángulo
A = sp.Matrix([x1, y1])
B = sp.Matrix([x2, y2])
C = sp.Matrix([x3, y3])

#Calcular los puntos medios de los lados
M_AB = (A + B) / 2
M_BC = (B + C) / 2
M_CA = (C + A) / 2

#Calcular las pendientes de las rectas que contienen las medianas
m_AB = (B[1] - A[1]) / (B[0] - A[0])
m_BC = (C[1] - B[1]) / (C[0] - B[0])
m_CA = (A[1] - C[1]) / (A[0] - C[0])

#Definir las ecuaciones de las medianas
ecuacion_mediana_AM_BC = sp.Eq((y1 - M_BC[1]) - m_AB * (x1 - M_BC[0]), 0)
ecuacion_mediana_BM_CA = sp.Eq((y2 - M_CA[1]) - m_BC * (x2 - M_CA[0]), 0)
ecuacion_mediana_CM_AB = sp.Eq((y3 - M_AB[1]) - m_CA * (x3 - M_AB[0]), 0)

#Encontrar la intersección de las medianas (centroide)
centroide = sp.solve((ecuacion_mediana_AM_BC, ecuacion_mediana_BM_CA, ecuacion_mediana_CM_AB), (x1, y1))

print("El punto de intersección de las medianas (centroide) es:", centroide)
```

```python
#DESARROLLO CON NUMPY +EJEMPLO
#Vértices del triángulo
A = np.array([1, 1])
B = np.array([3, 5])
C = np.array([7, 2])

#Función para encontrar el punto medio entre dos puntos
def punto_medio(p1, p2):
    return (p1 + p2) / 2

#Puntos medios
mid_AB = punto_medio(A, B)
mid_BC = punto_medio(B, C)
mid_CA = punto_medio(C, A)

#Calcular las medianas (líneas que conectan los vértices con los puntos medios opuestos)
mediana_A = np.vstack((A, mid_BC))
mediana_B = np.vstack((B, mid_CA))
mediana_C = np.vstack((C, mid_AB))

#Calcular el centroide del triángulo
centroide = (A + B + C) / 3

#Dibujar el triángulo
plt.plot([A[0], B[0], C[0], A[0]], [A[1], B[1], C[1], A[1]], 'b-')

#Dibujar las medianas
plt.plot(mediana_A[:, 0], mediana_A[:, 1], 'r--', label='Mediana desde A')
plt.plot(mediana_B[:, 0], mediana_B[:, 1], 'g--', label='Mediana desde B')
plt.plot(mediana_C[:, 0], mediana_C[:, 1], 'y--', label='Mediana desde C')

#Gráfico
plt.figure()
plt.plot(centroide[0], centroide[1], 'ro', label='Centroide')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Triángulo con Medianas y Centroide')
plt.legend()
plt.grid(True)
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

```python
#PUNTO B
def calcular_angulo(vector1, vector2):
    producto_punto = np.dot(vector1, vector2)
    magnitud_vector1 = np.linalg.norm(vector1)
    magnitud_vector2 = np.linalg.norm(vector2)
    cos_theta = producto_punto / (magnitud_vector1 * magnitud_vector2)
    angulo_rad = np.arccos(cos_theta)
    angulo_grados = np.degrees(angulo_rad)
    return angulo_grados

#Definimos los vectores base en R^3
i = np.array([1, 0, 0])
j = np.array([0, 1, 0])
k = np.array([0, 0, 1])

#Vectores en R^3
vector_a = np.array([1, 2, 3])
vector_b = np.array([4, 5, 6])
vector_c = np.array([3, 2, 1])
vector_d = np.array([6, 5, 4])
#Calcular el ángulo con cada uno de los vectores base
anguloa_con_i = calcular_angulo(vector_a, i)
anguloa_con_j = calcular_angulo(vector_a, j)
anguloa_con_k = calcular_angulo(vector_a, k)

angulob_con_i = calcular_angulo(vector_b, i)
angulob_con_j = calcular_angulo(vector_b, j)
angulob_con_k = calcular_angulo(vector_b, k)

anguloc_con_i = calcular_angulo(vector_c, i)
anguloc_con_j = calcular_angulo(vector_c, j)
anguloc_con_k = calcular_angulo(vector_c, k)

angulod_con_i = calcular_angulo(vector_d, i)
angulod_con_j = calcular_angulo(vector_d, j)
angulod_con_k = calcular_angulo(vector_d, k)
#Resultados
print("Ángulo entre el vector a y el vector i:", anguloa_con_i, "grados")
print("Ángulo entre el vector a y el vector j:", anguloa_con_j, "grados")
print("Ángulo entre el vector a y el vector k:", anguloa_con_k, "grados")

print("Ángulo entre el vector b y el vector i:", angulob_con_i, "grados")
print("Ángulo entre el vector b y el vector j:", angulob_con_j, "grados")
print("Ángulo entre el vector b y el vector k:", angulob_con_k, "grados")

print("Ángulo entre el vector c y el vector i:", anguloc_con_i, "grados")
print("Ángulo entre el vector c y el vector j:", anguloc_con_j, "grados")
print("Ángulo entre el vector c y el vector k:", anguloc_con_k, "grados")

print("Ángulo entre el vector d y el vector i:", angulod_con_i, "grados")
print("Ángulo entre el vector d y el vector j:", angulod_con_j, "grados")
print("Ángulo entre el vector d y el vector k:", angulod_con_k, "grados")
```

```python
#PUNTO H
#Función para calcular el ángulo entre dos vectores en grados
def angulo_entre_vectores(v1, v2):
    cos_theta = np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2))
    angulo_rad = np.arccos(np.clip(cos_theta, -1.0, 1.0))
    angulo_grados = np.degrees(angulo_rad)
    return angulo_grados

#Vectores en R^3
vector_a = np.array([1, 2, 3])
vector_b = np.array([4, 5, 6])
vector_c = np.array([3, 2, 1])
vector_d = np.array([6, 5, 4])

#Calcular el producto cruz entre a y b
producto_cruz_ab = np.cross(vector_a, vector_b)
#Calcular el producto cruz entre b y c
producto_cruz_bc = np.cross(vector_b, vector_c)
#Calcular el producto cruz entre c y d
producto_cruz_cd = np.cross(vector_c, vector_d)

#Calcular el ángulo entre el producto cruz y d
anguloab_con_d = angulo_entre_vectores(producto_cruz_ab, vector_d)
angulobc_con_d = angulo_entre_vectores(producto_cruz_bc, vector_d)
angulocd_con_d = angulo_entre_vectores(producto_cruz_cd, vector_d)
#Imprimir el producto cruz y el ángulo
print("El producto cruz entre a y b es:", producto_cruz_ab)
print("El producto cruz entre c y b es:", producto_cruz_bc)
print("El producto cruz entre c y d es:", producto_cruz_cd)
print("El ángulo entre el producto cruz de axb y d en R^3 es:", anguloab_con_d, "grados")
print("El ángulo entre el producto cruz de cxb y d en R^3 es:", angulobc_con_d, "grados")
print("El ángulo entre el producto cruz de cxd y d en R^3 es:", angulocd_con_d, "grados")
```

```python
#Definir el campo de fuerza F = <F_x, F_y>
def campo_de_fuerza(i, j):
    Fx = i**2
    Fy = j**2

    return np.array([Fx, Fy])

#Parámetros del circulo
radio = 1
inicio = 0
final = np.pi
num_puntos = 1000  # Número de puntos para la aproximación numérica al integrar

#Parametrización del círculo
t_valores = np.linspace(inicio, final, num_puntos)
x = radio * np.cos(t_valores)
y = radio * np.sin(t_valores)


#Derivada de la trayectoria
rx_dt = -radio * np.sin(t_valores)
ry_dt = radio * np.cos(t_valores)


#Evaluar el campo de fuerza en la trayectoria
Fx, Fy = campo_de_fuerza(x, y)

#Calcular el trabajo mediante producto punto vectorial
trabajo = np.sum(Fx * rx_dt + Fy * ry_dt) * (final - inicio) / num_puntos

print("El trabajo realizado por el campo de fuerza al moverse a lo largo del círculo es:", trabajo)
```
